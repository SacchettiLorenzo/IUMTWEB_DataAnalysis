```python
import numpy as np
import pandas as pd
import pymongo
```

```python
reviews_df = pd.read_csv('../../dataset/rotten_tomatoes_reviews.csv')
reviews_df
```

```python
reviews_df['rotten_tomatoes_link'] = reviews_df['rotten_tomatoes_link'].astype(pd.StringDtype())
reviews_df['movie_title'] = reviews_df['movie_title'].astype(pd.StringDtype())
reviews_df['critic_name'] = reviews_df['critic_name'].astype(pd.StringDtype())
reviews_df['publisher_name'] = reviews_df['publisher_name'].astype(pd.StringDtype())
reviews_df['review_type'] = reviews_df['review_type'].astype(pd.StringDtype())
reviews_df['review_score'] = reviews_df['review_score'].astype(pd.StringDtype())
reviews_df['review_date'] = pd.to_datetime( reviews_df['review_date'])
reviews_df['review_content'] = reviews_df['review_content'].astype(pd.StringDtype())
```

```python
reviews_df = reviews_df[~reviews_df.duplicated(subset=['rotten_tomatoes_link','critic_name','publisher_name','review_type','review_date','review_score', 'review_content'])].copy()
```

```python
reviews_df.to_csv('../../dataset/cleaned/rotten_tomatoes_reviews.csv', index=False)
```

```python
reviews_df = reviews_df.replace({np.nan: None})
```

```python
movies_grouped = []
for movie, group in reviews_df.groupby('movie_title'):
    reviews = group.apply(lambda row: {
        "critic_name": row['critic_name'],
        "top_critic": row['top_critic'],
        "publisher_name": row['publisher_name'],
        "review_type": row['review_type'],
        "review_score": row['review_score'],
        "review_date": row['review_date'],
        "review_content": row['review_content']
    }, axis=1).tolist()

    movies_grouped.append({
        "movie_title": movie,
        "reviews": reviews
    })
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["RT_reviews"]
mycol = mydb["Reviews"]
mycol.insert_many(movies_grouped)
```
